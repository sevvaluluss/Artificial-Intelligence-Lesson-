In this notebook, we will be demonstrating the GridLand environment.

Next, we will importing the `GridLand` class.

```python
from aitools.envs import GridLand
```

We create and display and instance of the `GridLand` class in the cell below. Experiment with changing the values of the the `size`, `hills`, and `random_state` parameters.

The `board` attribute of a `GridLand` instance is an array containing the costs associated with moving through each cell.  

```python
grid = GridLand(size=12, hills=16, random_state=37)
print(grid.board)
grid.display()
```

The `state` attribute contains the current location of our agent in the `GridLand` environment. The `goal` attribute contains the coordinates of the goal state.

```python
print(grid.state)
print(grid.goal)
```

The `get_actions()` method returns a list of actions available in the current state. The actions are encoded as follows:

0 - Move Up, 1 - Move Right, 2 - Move Down, 3 - Move Left

In the starting state, the only available options are moving right or down.

```python
print(grid.get_actions())
```

The `take_action()` method can be used to apply the results of an action. This method returns a new `GridLand` instance.

```python
grid = grid.take_action(2)
grid.display()
```

We will check the new state and see what actions are available in that state.

```python
print(grid.state)
print(grid.get_actions())
```

We will now recreate our `GridLand` instance and will provide our agent with a sequence of actions to take. Feel free to experiment with this sequence of actions.

```python
grid = GridLand(size=12, hills=16, random_state=37)
for a in [1, 1, 1, 1, 2, 2, 3, 0]:
    grid = grid.take_action(a)
grid.display()
```

The `get_path()` method returns a list of actions that have been taken, in the order in which they were taken.

```python
print(grid.get_path())
```

The `path_cost()` method returns the cost of the current path.

```python
print(grid.path_cost())
```

The `check_solved()` method checks to see if the current state is equal to the goal state.

```python
print(grid.check_solved())
```

The `soln_info()` method displays the length and cost of the current path.

```python
grid.soln_info()
```

We will now construct a path that is a valid solution.

```python
grid = GridLand(size=12, hills=16, random_state=37)

actions = [1] * 11 + [2] * 11
for a in actions:
    grid = grid.take_action(a)
grid.display()
```

```python
print(grid.check_solved())
print(grid.soln_info())
```
